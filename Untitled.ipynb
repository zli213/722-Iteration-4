```python
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler, OneHotEncoder, StringIndexer
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# # Renaming columns to remove spaces and special characters
# def sanitize_column_name(name):
#     return name.replace(' ', '_').replace('.', '_')

# # Sanitize column names
# for col_name in balanced_df_oversampled.columns:
#     sanitized_name = sanitize_column_name(col_name)
#     if sanitized_name != col_name:
#         balanced_df_oversampled = balanced_df_oversampled.withColumnRenamed(col_name, sanitized_name)

# Ensure the target column is correctly referenced
target_column = sanitize_column_name("4__ML/AI")
if target_column in balanced_df_oversampled.columns:
    balanced_df_oversampled = balanced_df_oversampled.withColumn("target", col(target_column))
else:
    raise ValueError(f"Column '{target_column}' does not exist in the DataFrame.")

# Split the dataset
train_data, test_data = balanced_df_oversampled.randomSplit([0.8, 0.2], seed=42)

# Create a pipeline for data processing
jobLevel_indexer = StringIndexer(inputCol='Job_Level', outputCol='JobLevelIndex', handleInvalid='keep')
jobLevel_encoder = OneHotEncoder(inputCol='JobLevelIndex', outputCol='JobLevelVec')

state_indexer = StringIndexer(inputCol='State', outputCol='StateIndex', handleInvalid='keep')
state_encoder = OneHotEncoder(inputCol='StateIndex', outputCol='StateVec')

country_indexer = StringIndexer(inputCol='Country', outputCol='CountryIndex', handleInvalid='keep')
country_encoder = OneHotEncoder(inputCol='CountryIndex', outputCol='CountryVec')

pipeline = Pipeline(stages=[
    jobLevel_indexer, jobLevel_encoder, 
    state_indexer, state_encoder, 
    country_indexer, country_encoder
])

# Fit the pipeline on training data
model = pipeline.fit(train_data)
encoded_train_df = model.transform(train_data)
encoded_test_df = model.transform(test_data)

# Get all columns except the categorical and target columns for the feature assembler
categorical_columns = ['Job_Level', 'State', 'Country']
encoded_columns = ['JobLevelVec', 'StateVec', 'CountryVec']
# Correctly exclude the 'target' column and the original target column
feature_cols = [col for col in encoded_train_df.columns if col not in categorical_columns + encoded_columns + ['target', target_column]]

# Use VectorAssembler to combine all features into a single vector
assembler = VectorAssembler(inputCols=feature_cols + encoded_columns, outputCol="features")

# Transform the datasets
final_train_df = assembler.transform(encoded_train_df)
final_test_df = assembler.transform(encoded_test_df)

# Initialize Logistic Regression model
lr = LogisticRegression(labelCol="target", featuresCol="features")

# Set parameter grid
param_grid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.01, 0.1, 1.0]) \
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0]) \
    .build()

# Set up CrossValidator
crossval = CrossValidator(estimator=lr,
                          estimatorParamMaps=param_grid,
                          evaluator=BinaryClassificationEvaluator(labelCol="target"),
                          numFolds=3)

# Train the model
cv_model = crossval.fit(final_train_df)

# Make predictions on the test set
predictions = cv_model.transform(final_test_df)

# Evaluate the model
evaluator = BinaryClassificationEvaluator(labelCol="target")
auc = evaluator.evaluate(predictions)
print(f"AUC: {auc}")

# Show part of the prediction results
predictions.select("features", "target", "prediction", "probability").show(truncate=False)

# Show specific columns of the prediction results
predictions.select('target', 'prediction').show()
```

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

# Create a list to store the silhouette scores for each K value
silhouette_scores = []

# Define the range of K values to evaluate
K_range = range(2, 11)  # You can adjust this range as needed

# Loop through each K value
for k in K_range:
    # Initialize KMeans with the current K value
    kmeans = KMeans(featuresCol='features', k=k, maxIter=10, seed=42)
    
    # Fit the model on the training data
    kmeans_model = kmeans.fit(train_tree_data)
    
    # Make predictions on the test set
    kmeans_predictions = kmeans_model.transform(test_tree_data)
    
    # Evaluate clustering by computing the Silhouette score
    evaluator = ClusteringEvaluator(featuresCol='features', metricName='silhouette', distanceMeasure='squaredEuclidean')
    silhouette = evaluator.evaluate(kmeans_predictions)
    
    # Store the Silhouette score
    silhouette_scores.append((k, silhouette))

    print(f"K = {k}, Silhouette Score = {silhouette}")

# Find the K value with the highest Silhouette score
best_k, best_silhouette = max(silhouette_scores, key=lambda x: x[1])
print(f"\nBest K: {best_k} with Silhouette Score: {best_silhouette}")

# Display the silhouette scores for different K values
for k, silhouette in silhouette_scores:
    print(f"K = {k}, Silhouette Score = {silhouette}")

# Finally, fit the KMeans model with the best K value
kmeans = KMeans(featuresCol='features', k=best_k, maxIter=10, seed=42)
kmeans_model = kmeans.fit(train_tree_data)

# Make predictions
kmeans_predictions = kmeans_model.transform(test_tree_data)

# Display the cluster centers
centers = kmeans_model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)
```
